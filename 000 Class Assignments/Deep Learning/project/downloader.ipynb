```python
from icrawler.builtin import GoogleImageCrawler as Crawler
```

```python
aircraft_types = ['bomber plane']
```

```python
while aircraft_types:
    word = aircraft_types.pop()
    crawler = Crawler(storage={'root_dir': word},
                      feeder_threads=1,
                      parser_threads=2,
                      downloader_threads=4
                      )
    crawler.crawl(keyword=word, max_num=200)
```
