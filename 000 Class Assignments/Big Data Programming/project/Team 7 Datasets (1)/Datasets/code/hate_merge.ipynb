```python
import pandas as pd
import numpy as np
import os
```

```python
main_excel = pd.read_csv('hate_crime.csv')
```

```python
df = main_excel.copy()
```

```python
main_excel.columns
keep_cols = ['PUB_AGENCY_NAME', 'AGENCY_TYPE_NAME', 'STATE_NAME', 'OFFENSE_NAME', 'TOTAL_OFFENDER_COUNT']
drop_cols = [cols for cols in list(main_excel.columns) if cols not in keep_cols]

df.drop(drop_cols, axis=1, inplace=True)
```

```python
df['TOTAL_OFFENDER_COUNT'].replace(0, 1, inplace=True)
df
```

```python
index_cols = ['PUB_AGENCY_NAME', 'AGENCY_TYPE_NAME', 'STATE_NAME', 'TOTAL_OFFENDER_COUNT']

df = (df
.set_index(index_cols)
.apply(lambda x: x.str.split(';').explode())
)
df.head(25)
```

```python
df = df.reset_index()
df
```

```python
group_cols = ['PUB_AGENCY_NAME', 'AGENCY_TYPE_NAME', 'STATE_NAME', 'OFFENSE_NAME']
df = (df
.groupby(group_cols)
.sum()
.reset_index()
)
df
```

```python
try:
    os.mkdir(f'{os.getcwd()}\\hate_output')
except FileExistsError:
    pass
```

```python
new_col_names = ['Agency Name', 'Agency Type', 'State', 'Offense Name', 'Total Offenses']
df.to_excel(f'{os.getcwd()}\\hate_output\\hate_crime_cleaned.xlsx', na_rep = 'NA', sheet_name='Cities', header=new_col_names, index=False)
```
