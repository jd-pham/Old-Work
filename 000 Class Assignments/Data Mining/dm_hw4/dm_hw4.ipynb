```python
import pandas as pd
import graphviz
 
from sklearn import tree

import math
 
df1 = pd.read_excel('dataset.xlsx', sheet_name=0, index_col=0)
df2 = pd.read_excel('dataset.xlsx', sheet_name=1)
 
df1 = df1.replace(['No'], 0).replace(['Yes'], 1)
cols = df1.columns[:-1]
 
features = df1[cols]
target = df1['Grass']
 
clf = tree.DecisionTreeClassifier()
clf = clf.fit(df1[cols], target)
 
graph = tree.export_graphviz(clf, feature_names=cols,
                           class_names=['Dry', 'Wet'],
                           filled=True, rounded=True,
                           special_characters=True)
 
graph = graphviz.Source(graph)
graph
 
 
 
df2 = pd.read_excel('dataset.xlsx', sheet_name=1, index_col=0)
 
df2 = df2.replace(['No'], 0).replace(['Yes'], 1)
cols = df2.columns[:-1]
 
features = df2[cols]
target = df2['Grass']
 
clf = tree.DecisionTreeClassifier()
clf = clf.fit(df2[cols], target)

graph = tree.export_graphviz(clf, feature_names=cols,
                           class_names=['Dry', 'Wet'],
                           filled=True, rounded=True,
                           special_characters=True)
 
graph = graphviz.Source(graph)
graph
 
 

```
