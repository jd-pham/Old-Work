```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('hmeq.csv')
df
```

```python
q1 = df.dropna()
q1
```

```python
q1_filled = df.__deepcopy__()
for col in df.columns:
    try:
        mean = df[col].mean()
    except(TypeError):
        continue

    q1_filled[col].fillna(mean, inplace=True)
q1_filled
    
```

```python
df['LOAN'].mean()
```

```python

def find_means(df):
    for col in df.columns:
        try:
            mean = df[col].mean()
        except(TypeError):
            continue

        replaced_means[col] = mean 
    return replaced_means
```

```python
not_filled_means = find_means(q1_filled)
filled_means = find_means(q1)


print(not_filled_means)
print()
print(filled_means)
```
